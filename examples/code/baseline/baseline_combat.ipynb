```python
from neuroCombat import neuroCombat
import sys
sys.path.append("/DATA/WGS_study/YSL/projects/latentgee/examples/code")

from functions import evaluate_batch_correction

data = X_filtered_df3.T.values  # features x samples
batch = batch_filtered_df3.values
covars = {'hivstatus': meta_filtered_df3['hivstatus'].values,
          'Age': meta_filtered_df3['Age'].values,
          'gender': meta_filtered_df3['gender'].values,
          'batch': meta_filtered_df3['Study'].values}
covars = pd.DataFrame(covars)
data_combat = neuroCombat(dat=data, 
                          covars=covars,
                          batch_col='batch')

# 보정된 데이터 추출 (features x samples → samples x features로 전치)
X_corrected_combat_df3 = pd.DataFrame(
    data_combat['data'].T,
    index=X_filtered_df3.index,
    columns=X_filtered_df3.columns
)

print(f"shape: {X_corrected_combat_df3.shape}")
print(f"max: {X_corrected_combat_df3.values.max():.4f}")
print(f"min: {X_corrected_combat_df3.values.min():.4f}")
# ComBat df3 평가


combat_df3_result = evaluate_batch_correction(
    X_before      = X_filtered_df3.values,
    X_after       = X_corrected_combat_df3.values,
    batch_labels  = batch_filtered_df3['Study'].values,
    bio_labels    = meta_filtered_df3['hivstatus'].values,
    renormalize   = True,
    label         = "df3 — ComBat"
)

X_corrected_combat_df3.to_csv("/DATA/WGS_study/YSL/projects/latentgee/examples/results/baseline_results/X_df3_corrected_ComBat.csv")
```
